```python
import pandas as pd
import urllib.request

# Charger le dataset
df = pd.read_excel('datasets\Lunettes_.xlsx', index_col=0)

```

```python
# Mapping des styles à leurs traductions
style_mapping = {
    'Browline': 'Ligne Des Sourcils',
    'Style Enveloppante': 'Enveloppante',
    'Aviator': 'Aviateur',
    'Gradient Blue': 'Ligne Des Sourcils',
    'Transparent': 'Ligne Des Sourcils',
    'Gris': 'Aviateur',
    'Amber Yellow': 'Aviateur',
    'Marron': 'Enveloppante',
    'Light Amber Brown': 'Aviateur',
    'Amber Brown': 'Carrée',
    'Gradient Grey': 'Carrée'
}

# Remplacer les styles par leur traduction
df['Style'] = df['Style'].replace(style_mapping)

```

```python
# Dictionnaire de chemins de fichiers par style
file_paths = {
    'Aviateur': 'images/Style/Aviateur/',
    'Papillon': 'images/Style/Papillon/',
    'Rectangulaire': 'images/Style/Rectangulaire/',
    'Enveloppante': 'images/Style/Enveloppante/',
    'Ovale': 'images/Style/Ovale/',
    'Carrée': 'images/Style/Carrée/',
    'Oeil de Chat': 'images/Style/Oeil de Chat/',
    'Ronde': 'images/Style/Ronde/',
    'Wayfarer': 'images/Style/Wayfarer/',
    'Oversize': 'images/Style/Oversize/',
    'Géométriques': 'images/Style/Géométriques/',
    'Verre Unique': 'images/Style/Verre Unique/',
    'Ligne Des Sourcils': 'images/Style/Ligne Des Sourcils/',
    'Masque': 'images/Style/Masque/',
    'Cerclée': 'images/Style/Cerclée/'
}

```

```python
# Fonction pour télécharger les images
def url_to_jpg(urls, file_path):
    for i, url in enumerate(urls):
        filename = f'image-{i}.jpg'
        full_path = f'{file_path}{filename}'
        urllib.request.urlretrieve(url, full_path)
        print(f'{filename} saved.')
```

```python
# Télécharger les images pour chaque style
for style, path in file_paths.items():
    urls = df['Image_URL-src'][df['Style'] == style].tolist()
    url_to_jpg(urls, path)
```
